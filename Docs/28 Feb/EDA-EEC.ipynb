```python
# Required Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset from seaborn library (you can also download it manually from Kaggle)
titanic = sns.load_dataset('titanic')

# Show first few rows of the dataset
print(titanic.head())

# Basic information about the dataset
print(titanic.info())

# Descriptive statistics
print(titanic.describe())

# Check for missing values
print(titanic.isnull().sum())

# Data Cleaning: Handle missing values (e.g., fill missing 'age' with mean)
titanic['age'].fillna(titanic['age'].mean(), inplace=True)
titanic['embarked'].fillna(titanic['embarked'].mode()[0], inplace=True)

# Data Cleaning: Remove duplicates if any
titanic.drop_duplicates(inplace=True)

# Exploratory Data Analysis: Visualizations
# 1. Distribution of Age
plt.figure(figsize=(8,6))
sns.histplot(titanic['age'], kde=True)
plt.title('Age Distribution')
plt.show()

# 2. Survival Rate by Gender
plt.figure(figsize=(8,6))
sns.countplot(data=titanic, x='sex', hue='survived')
plt.title('Survival Rate by Gender')
plt.show()

# 3. Survival Rate by Passenger Class
plt.figure(figsize=(8,6))
sns.countplot(data=titanic, x='pclass', hue='survived')
plt.title('Survival Rate by Passenger Class')
plt.show()

# 4. Age vs Fare (Relationship between Age and Fare)
plt.figure(figsize=(8,6))
sns.scatterplot(data=titanic, x='age', y='fare', hue='survived')
plt.title('Age vs Fare')
plt.show()

# 5. Correlation Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(titanic.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Insights: Calculate the survival rate
survival_rate = titanic['survived'].mean()
print(f"Overall Survival Rate: {survival_rate * 100:.2f}%")
```

```python
# Required Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset from seaborn library (you can also download it manually from Kaggle)
titanic = sns.load_dataset('titanic')

# Show first few rows of the dataset
print(titanic.head())

# Basic information about the dataset
print(titanic.info())

# Descriptive statistics
print(titanic.describe())

# Check for missing values
print(titanic.isnull().sum())

# Data Cleaning: Handle missing values with different methods
# Fill missing 'age' with median (since it’s a numerical column and may have skewed distribution)
titanic['age'].fillna(titanic['age'].median(), inplace=True)

# Fill missing 'embarked' with mode (most frequent value), as it's categorical
titanic['embarked'].fillna(titanic['embarked'].mode()[0], inplace=True)

# For 'fare', fill with mean (as it's a numerical column)
titanic['fare'].fillna(titanic['fare'].mean(), inplace=True)

# Data Cleaning: Remove duplicates if any
titanic.drop_duplicates(inplace=True)

# Exploratory Data Analysis: Visualizations
# 1. Distribution of Age
plt.figure(figsize=(8,6))
sns.histplot(titanic['age'], kde=True)
plt.title('Age Distribution')
plt.show()

# 2. Survival Rate by Gender
plt.figure(figsize=(8,6))
sns.countplot(data=titanic, x='sex', hue='survived')
plt.title('Survival Rate by Gender')
plt.show()

# 3. Survival Rate by Passenger Class
plt.figure(figsize=(8,6))
sns.countplot(data=titanic, x='pclass', hue='survived')
plt.title('Survival Rate by Passenger Class')
plt.show()

# 4. Age vs Fare (Relationship between Age and Fare)
plt.figure(figsize=(8,6))
sns.scatterplot(data=titanic, x='age', y='fare', hue='survived')
plt.title('Age vs Fare')
plt.show()

# 5. Correlation Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(titanic.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Insights: Calculate the survival rate
survival_rate = titanic['survived'].mean()
print(f"Overall Survival Rate: {survival_rate * 100:.2f}%")
```
